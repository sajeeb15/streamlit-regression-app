```python
import pandas as pd
import folium
from folium import IFrame
from google.colab import drive
```

```python
drive.mount('/content/drive')
```

```python
geospatial_data = pd.read_csv("/content/drive/MyDrive/Earthquakes.csv")
```

```python
geospatial_data_map = geospatial_data.head(2000)
```

```python
latitude = geospatial_data_map['location.latitude'].mean()
longitude = geospatial_data_map['location.longitude'].mean()
```

```python
m = folium.Map(location=[latitude, longitude], zoom_start=5)
```

```python
for index, row in geospatial_data_map.head(500).iterrows():
    popup_content = f"""
    Location: {row['location.full']}<br>
    Magnitude: {row['impact.magnitude']}<br>
    Significance: {row['impact.significance']}
    """
    iframe = IFrame(popup_content, width=250, height=100)
    popup = folium.Popup(iframe, max_width=300)

    folium.Marker(
        location=[row['location.latitude'], row['location.longitude']],
        popup=popup,
        icon=folium.Icon(color='red')
    ).add_to(m)

# 7. DISPLAY MAP DIRECTLY IN COLAB
m
```

```python

import folium
from folium.plugins import HeatMap

# 1. Data preparation
geospatial_data_map = geospatial_data.head(2000)
latitude = geospatial_data_map['location.latitude'].mean()
longitude = geospatial_data_map['location.longitude'].mean()

# 2. Folium Map create (basic map, navigation controls minimun thakbe, eta folium er default, eta off kora jay na)
m = folium.Map(location=[latitude, longitude], zoom_start=5, control_scale=False)  # Control scale off kora holo

# 3. Heatmap er jonno data prepare koro (latitude, longitude list)
heat_data = geospatial_data_map[['location.latitude', 'location.longitude']].values.tolist()

# 4. HeatMap add koro map e
HeatMap(heat_data, radius=15, blur=10).add_to(m)

# 5. Map DIRECT COLAB e display
m
```
